```python
from sklearn import linear_model
```

```python
scaler = StandardScaler().fit(X1)
features = scaler.transform(X1)

X1_std = pd.DataFrame(features, columns = X1.columns)
```

```python
# Train Lasso
alphas = 10**np.linspace(10,-2,500)*0.5
scores = np.empty_like(alphas)
cos = []
for i,a in enumerate(alphas):
    lasso = linear_model.Lasso()
    lasso.set_params(alpha=a)
    lasso.fit(X1_std, Y)
    scores[i] = lasso.score(X1_std, Y)
    cos.append(lasso.coef_)
```

```python
alphadf = pd.DataFrame({"alpha":alphas, "score":scores})
alphadf[alphadf.score==max(alphadf.score)]
```

```python
lassocv = linear_model.LassoCV(alphas = alphas, random_state=0)
lassocv.fit(X1_std, Y)
lassocv_score = lassocv.score(X1_std, Y)
lassocv_alpha = lassocv.alpha_
```

```python
print("alpha: ",lassocv_alpha)
print("R**2:",lassocv_score)
```

```python

lassocv_coef = pd.DataFrame({"Feature": X1.columns, "Coeff": lassocv.coef_})

lassocv_coef = lassocv_coef.sort_values(by=['Coeff'], ascending = False, key = abs)
lassocv_coef
```

```python
y_pred = lassocv.predict(X1_std)

RMSE = mean_squared_error(Y,y_pred, squared = False)
score = lassocv.score(X1_std,Y)

print(RMSE)
print(score)
```
