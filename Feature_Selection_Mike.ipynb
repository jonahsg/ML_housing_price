```python
def transform(DF):
    df_T = pd.DataFrame()
    df_T['LogSalePrice'] = train['LogSalePrice']
    for feature in DF.columns:
        X_T = DF.loc[:, feature]
        lam = boxcox_normmax(X_T + 1)
        print(f'lambda of {feature} = {lam}')
        tt = pd.DataFrame()
        tt['LogSalePrice'] = train['LogSalePrice']
        tt['SalePrice'] = Y
        tt.loc[:, feature] = boxcox1p(X_T, lam)
        df_T = pd.concat([df_T, tt.loc[:, feature]], axis = 1)
        print(tt.plot.scatter(x = feature, y= 'LogSalePrice'))
        print(train.plot.scatter(x = feature, y= 'LogSalePrice'))
    return df_T
    
```

```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)

calc_vif(X1)
```
